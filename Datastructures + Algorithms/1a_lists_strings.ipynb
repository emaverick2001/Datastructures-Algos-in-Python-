## .append(x)

  - Add an item to the end of a list
  - Similar to a[len(a):] = [x]

```python
# append to end - O(1)

fruits.append('grape')
print(fruits)
```

## .extend(iterable)

- Extend the list by appending all the items from the iterable. 
- Similar to a[len(a):] = iterable.

```python
# Concatenate many at once - O(k)
fruits.extend(['x','y'])
print(fruits)
```

## .insert(i,x)

- Insert an item at a given position in a list.
- i - index of the element before which to insert


```python
# append to front - O(n)

fruits.insert(0,'passionfruit')
print(fruits)
```

```python
# append within the list - O(n) since have to shift all elements to the right to maintain size of array

fruits.insert(2,'plum')
print(fruits)
```

```python
# append to end - O(1)

fruits.insert(len(fruits),'mango')
print(fruits)
```

## .remove(x)

- Remove the first item from the list whose value is equal to x

```python
# remove within list - O(n) since have to shift all elements to left to maintain size of array
# remove at end - O(1)

fruits.remove('banana')
print(fruits)
```

## .index(x)

- Return zero-based index of the first occurrence of x in the list. 
- Raises a ValueError if there is no such item.

```python
print(fruits.index('pear'))
print(fruits.index('banana'))
```

## .strip()

```python
"""
### .strip() / .lstrip() / .rstrip()

- Remove leading/trailing whitespace by default.
- Provide a character set to strip specific chars.
"""
# (Code)
t = "   \t  hello  \n"
print(t.strip())      # 'hello'
print(t.lstrip())     # 'hello  \n'
print(t.rstrip())     # '   \t  hello'
u = "~~--data--~~"
print(u.strip("~-"))  # 'data--' (strips any of '~' or '-' from ends)
```

## .split()

```python
"""
### .split() / .rsplit()

- split() with no sep splits on any whitespace and collapses runs.
- sep specifies exact delimiter; maxsplit controls number of splits.
- rsplit() splits from the right.
"""

line = "a  b\tc\n\n d"
print(line.split())             # ['a', 'b', 'c', 'd']
csv = "a,b,c,d"
print(csv.split(","))           # ['a', 'b', 'c', 'd']
print(csv.split(",", maxsplit=2))   # ['a', 'b', 'c,d']
print(csv.rsplit(",", maxsplit=1))  # ['a,b,c', 'd']
```

## Case Methods

```python
"""
## Case Methods

- .upper(), .lower(), .title(), .capitalize(), .casefold() (stronger lower for caseless matching).
"""
# (Code)
s = "Straße"
print(s.lower())    # 'straße'
print(s.casefold()) # 'strasse' (better for comparisons)
print("hello world".title())     # 'Hello World'
print("hello world".capitalize())# 'Hello world'
```

## Count & Predicates

```python
"""
## Count & Predicates

- .count(sub) occurrences.
- .isalpha(), .isalnum(), .isdigit(), .isdecimal(), .isnumeric(), .isspace(), .istitle().
"""
# (Code)
print("mississippi".count("ss"))   # 2
print("abc".isalpha(), "abc123".isalnum(), "123".isdigit())
print("Ⅳ".isnumeric(), "١٢٣".isdecimal())  # True False (unicode digits nuance)
print("Hello World".istitle())     # True
```

## startswith() / endswith()

```python
"""
## startswith() / endswith()

- Accept tuple of prefixes/suffixes.
"""
# (Code)
fn = "report.csv"
print(fn.endswith((".csv", ".tsv")))   # True
print("http://example.com".startswith(("http://", "https://")))  # True
```

## partition() / rpartition()

```python
"""
## partition() / rpartition()

- Split into (head, sep, tail) at first/last occurrence; always returns 3-tuple.
"""
# (Code)
print("key=value".partition("="))   # ('key', '=', 'value')
print("a=b=c".rpartition("="))      # ('a=b', '=', 'c')
```
