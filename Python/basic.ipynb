### Indexing

```python
# extracts character @ specified index
word = 'Python'
print(word[0])
print(word[5])

print(word[-1])
```

### Slicing

```python
# characters from position 0 (included) to 2 (excluded)
word = 'Savestate'
word[0:2]
```

```python
# character from the beginning to position 2 (excluded)
print(word[:2])

# characters from position 4 (included) to the end
print(word[4:])

# characters from the second-last to the end
print(word[-2:])
```

## for Statements

```python
# looping through a list
words = ['cat', 'window', 'defenestrate']
for w in words:
    print(w, len(w))
```

```python
# looping through a dictionary
knights = {'gallahad': 'the pure', 'robin': 'the brave'}
for k, v in knights.items():
    print(k, v)
```

### range()

```python
for i in range(5):
    print(i)
```

```python
for i in range(0,10,2):
    print(i)
```

### enumerate()

- gives (index, value) pairs

```python
for i, v in enumerate(['tic', 'tac', 'toe']):
    print(i, v)
```

### zip()

```python
questions = ['name', 'quest', 'favorite color']
answers = ['lancelot', 'the holy grail', 'blue']
for q, a in zip(questions, answers):
    print(f'What is your {q}?  It is {a}.')
```

### break / continue / else

```python
for num in [2, 3, 5, 7, 9]:
    if num % 2 == 0:
        print("even", num)
        break
else:
    print("no even found")  # executes only if loop *didnâ€™t* break

# continue skips current iteration
for i in range(5):
    if i == 2:
        continue
    print(i)
```

### Comprehensions

- Equivalent to building new lists inline

```python
squares = [x**2 for x in range(5)]
print(squares)
```

```python
# Conditional comprehension
even_squares = [x**2 for x in range(10) if x%2==0]
print(even_squares)
```

```python
# Dict and set comprehensions
d = {x: x**2 for x in range(3)}   # {0:0,1:1,2:4}
s = {x for x in "abracadabra" if x not in "abc"}
```
