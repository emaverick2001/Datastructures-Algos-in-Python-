## VSCODE Keyboard Shortcuts

- https://code.visualstudio.com/shortcuts/keyboard-shortcuts-macos.pdf
- https://www.notion.so/VS-Code-Quick-Commands-21fb552a8bcf80f79efefd0bd2dc3073?source=copy_link

### General

#### cmd + ,

- open user settings

### Navigation

### option + > 

- skip words in a line of text

### cmd + >

  - go to the end of a line

### cmd + o

- open a file in vscode

### cmd + shift + o

- Jump to a function or class within a file

### cmd + click


- Quickly view where a variable/function is defined

### ctrl + g

- jump to a specific line number

### cmd + p

- Type part of a filename to open it instantly

### shift + cmd + p

- quickly access vscode searchbar for reloading vscode window + other settings

### cmd + n or ctrl + opt + cmd + n

- create a new file
